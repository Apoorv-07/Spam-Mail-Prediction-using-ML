Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

Data Collection and Preprocessing

```python
#reading the csv file
csv_data = pd.read_csv('/content/mail_data.csv')
print(csv_data)
```

```python
#replace null values to null strings
fixed_csv = csv_data.where(pd.notnull(csv_data),'')

#Converting cleaned csv to dataframe
df = pd.DataFrame(fixed_csv)
df.head()
```

```python
df.shape
```

Label Encoding </br>
spam -> 0 </br>
ham -> 1

```python
df.loc[df['Category'] == 'spam','Category'] = 0
df.loc[df['Category'] == 'ham','Category'] = 1
```

```python
df.head()
```

```python
#seperating data as texts and labels
X = df['Message']
Y = df['Category']
```

```python
print(X)
```

```python
print(Y)
```

Splitting Data

```python
X_train , X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2,random_state=3)
```

Feature Extraction

```python
#transform the text data into feature vectors that can be used as input for Logistic Regression Model
feature_extractor = TfidfVectorizer(min_df=1 , stop_words = 'english', lowercase=True)


X_train_features = feature_extractor.fit_transform(X_train)
X_test_features = feature_extractor.transform(X_test)

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

Training the Logistic Regression Model with this data

```python
#Loading Model
model = LogisticRegression()
```

```python
#Training the model
model.fit(X_train_features,Y_train)
```

Evaluating the Trained Model

```python
Y_training_pred = model.predict(X_train_features)
train_score = accuracy_score(Y_train,Y_training_pred)
print(train_score)
```

```python
Y_pred = model.predict(X_test_features)
test_score = accuracy_score(Y_test,Y_pred)
print(test_score)
```

Building a Predictive System

```python
input_mail = ["Aft i finish my lunch then i go str down lor. Ard 3 smth lor. U finish ur lunch already?"]
input_features = feature_extractor.transform(input_mail)

pred = model.predict(input_features)
if(pred == 0):
  print("SPAM MAIL")
else:
  print("Ham Mail")
```
